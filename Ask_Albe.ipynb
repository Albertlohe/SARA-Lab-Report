## üìö Step 3: Import Libraries

```python
from groq import Groq
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

print('‚úÖ Libraries imported successfully!')
```

---
## ü§ñ Step 4: Set Up Albe's Personality

```python
# Connect to Groq
client = Groq(api_key=GROQ_API_KEY)

# Albe's personality ‚Äî you can customize this!
SYSTEM_PROMPT = """
You are Albe, a friendly, smart, and helpful AI assistant.
You give clear, concise, and accurate answers.
You are conversational and easy to understand.
If you don't know something, you say so honestly.
"""

# Chat history memory
chat_history = []

print('‚úÖ Albe is ready!')
```

---
## üí¨ Step 5: Chat with Albe!

Run this cell to start chatting. Type your message and press **Send** or hit **Enter**!

```python
# ---- UI Components ----
chat_display = widgets.Output(
    layout=widgets.Layout(
        border='1px solid #ddd',
        min_height='350px',
        max_height='400px',
        overflow_y='auto',
        padding='10px',
        background_color='#f9f9f9',
        border_radius='10px'
    )
)

text_input = widgets.Text(
    placeholder='Ask Albe anything...',
    layout=widgets.Layout(width='75%', height='40px')
)

send_button = widgets.Button(
    description='Send üì®',
    button_style='primary',
    layout=widgets.Layout(width='20%', height='40px')
)

clear_button = widgets.Button(
    description='Clear Chat üóëÔ∏è',
    button_style='warning',
    layout=widgets.Layout(width='96%', height='35px')
)

status_label = widgets.Label(value='')

# ---- Display Header ----
display(HTML("""
<div style='text-align:center; padding:15px; background:linear-gradient(135deg, #667eea, #764ba2);
border-radius:12px; margin-bottom:10px;'>
    <h2 style='color:white; margin:0;'>ü§ñ Ask Albe</h2>
    <p style='color:#ddd; margin:4px 0 0;'>Your personal AI assistant</p>
</div>
"""))

display(chat_display)
display(widgets.HBox([text_input, send_button]))
display(clear_button)
display(status_label)

# ---- Chat Function ----
def send_message(event=None):
    user_msg = text_input.value.strip()
    if not user_msg:
        return

    text_input.value = ''
    status_label.value = '‚è≥ Albe is thinking...'

    # Add to history
    chat_history.append({"role": "user", "content": user_msg})

    # Show user message
    with chat_display:
        display(HTML(f"""
        <div style='text-align:right; margin:8px 0;'>
            <span style='background:#667eea; color:white; padding:8px 14px;
            border-radius:18px 18px 4px 18px; display:inline-block;
            max-width:75%; font-size:14px;'>üë§ {user_msg}</span>
        </div>
        """))

    # Get AI response
    try:
        messages_with_system = [{"role": "system", "content": SYSTEM_PROMPT}] + chat_history
        response = client.chat.completions.create(
            model="llama3-8b-8192",
            messages=messages_with_system
        )
        reply = response.choices[0].message.content
        chat_history.append({"role": "assistant", "content": reply})

        with chat_display:
            display(HTML(f"""
            <div style='text-align:left; margin:8px 0;'>
                <span style='background:#f0f0f0; color:#333; padding:8px 14px;
                border-radius:18px 18px 18px 4px; display:inline-block;
                max-width:75%; font-size:14px;'>ü§ñ {reply}</span>
            </div>
            """))
        status_label.value = '‚úÖ Done!'

    except Exception as e:
        status_label.value = f'‚ùå Error: {str(e)}'

# ---- Clear Function ----
def clear_chat(event=None):
    chat_history.clear()
    with chat_display:
        clear_output()
    status_label.value = 'üóëÔ∏è Chat cleared!'

# ---- Connect Buttons ----
send_button.on_click(send_message)
clear_button.on_click(clear_chat)
text_input.on_submit(send_message)
```
